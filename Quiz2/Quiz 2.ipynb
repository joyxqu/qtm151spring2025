Step 1:
- Define a function with a parameter "TotalStats"
- If "TotalStats > 85", classification = "Top Performer"
- If "TotalStats <= 85", classification = "Non Top-Performer"
- The function returns "classification"

Step 2:
- Extract the column "TotalStats" in the dataset <br>
- Create a new variable called "player_classification" <br>
to assign each player to the correct classification

HINT: Use the ".apply()" function

````python
# Write your own code
# Define fcn_performance with parameter 'TotalStats'
def fcn_performance(TotalStats):

# If 'TotalStats' > 85 print "Top Performer"
    if TotalStats > 85:
        classification = "Top Performer"

# Else print "Non Top-Performer"
    else:
        classification = "Non Top-Performer"

# Return the value
    return(classification)

# Display the 'TotalStats' column
display(fifa23['TotalStats'])

# Create a new column 'player_classification'
# Apply fcn_performance to 'TotalStats'
fifa23['player_classification'] = fifa23['TotalStats'].apply(fcn_performance)

# Display the new 'player_classification' column
display(fifa23['player_classification'])
````

(d) Regression function

<font size = "3">

- Run a regression of "Overall" (left-hand side) on "Age" (right-hand side)
- What are the values of the coefficients?

[HINT: Use the commands ```smf.ols()``` and ```.fit()``` and ```.params``` <br>
from Lecture 11 ]


````python
# Write your own code
# Run a regression of 'Overall ~ Age'
model = smf.ols(formula = 'Overall ~  Age', data = fifa23)

# Fit the model with standard errors 'cov'
results = model.fit()

# Retrieve the parameters from the model
b_list = results.params

# Print the coefficient values
print(b_list)
````

The values of the coefficients for the intercept and slope are 49.787345 and 0.636467 respectively.

(e) Split a dataset into subsets

<font size = "3">

- Store the total number of rows of the dataset in a variable called ```n``` <br>
- Create a new variable called ```random_var``` drawn from a <br>
uniform distribution between 0 and 1 with size ```n```
- Use .query() to subset all observations with ```random_var <0.4```.<br>
store it in a new dataset called "data0"
- Use .query() to subset all observations with ```random_var >= 0.4```.<br>
store it in a new dataset called "data1"

[HINT: See Lecture 2 for how to generate a uniform random variable]


````python
# Write your own code
# Count the number of rows in 'fifa23'
n = len(fifa23)

# Create a 'random_var' of uniform distribution between 0 and 1 with size n
fifa23['random_var'] = np.random.uniform(low = 0,high = 1, size = n)

# Subset 'fifa23' for 'random_var < 0.4'
data0 = fifa23.query('random_var < 0.4')

# Subset 'fifa23' for 'random_var >= 0.4'
data1 = fifa23.query('random_var >= 0.4')
````
