```python
# Load the China CO2 emissions data
china_emissions = pd.read_csv('data/china_co2_emissions.csv')

# Load the Germany CO2 emissions data
germany_emissions = pd.read_csv('data/germany_co2_emissions.csv')

# Plot a scatter plot of the China CO2 emissions data
# Add year to the x-axis and total emissions to the y-axis
plt.scatter(x = china_emissions['year'], y = china_emissions['total_emissions'])

# Add a title and labels
plt.title("China's CO2 Emissions Over Time")
plt.xlabel("Year")
plt.ylabel("Total Emissions")

# Show the plot
plt.show()

# Plot a scatter plot of the Germany CO2 emissions data
# Add year to the x-axis and total emissions to the y-axis
plt.scatter(x = germany_emissions['year'], y = germany_emissions['total_emissions'])

# Add a title and labels
plt.title("Germany's CO2 Emissions Over Time")
plt.xlabel("Year")
plt.ylabel("Total Emissions")

# Show the plot
plt.show()
```
